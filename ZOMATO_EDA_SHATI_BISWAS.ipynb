## 4. Visualize missing values

* Heatmap of missing values:

```python
# Visualizing null values with a heatmap
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cmap='viridis', cbar=False)
plt.title('Heatmap of Missing Values')
plt.show()
```

* Bar plot of missing values:

```python
# Calculate the percentage of missing values for each column
missing_percentage = df.isnull().mean() * 100

# Plot the missing data as a bar plot
missing_percentage.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Percentage of Missing Values by Column')
plt.ylabel('Percentage')
plt.show()
```

* Grid of missing values:

```python
import numpy as np

# Replace missing values with a distinct color in a grid
plt.imshow(df.isnull(), cmap="cividis", aspect='auto', interpolation='none')
plt.colorbar(label='Missing Data (1 = Missing, 0 = Present)')
plt.title('Grid of Missing Values')
plt.show()
```

# remove outlier from average_cost_for_two

```python
Q1 = final_df['Average Cost for two'].quantile(0.25)
Q3 = final_df['Average Cost for two'].quantile(0.75)
IQR = Q3 - Q1

outliers = final_df[(final_df['Average Cost for two'] < (Q1 - 1.5 * IQR)) | (final_df['Average Cost for two'] > (Q3 + 1.5 * IQR))]
print(outliers)
```

# multivariate analysis

### Correlation Matrix for Numerical Variables:
To check relationships between numerical variables, use the correlation matrix.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation matrix
corr_matrix = final_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

#### Observation: This correlation matrix highlights the strength and direction of relationships between numerical variables (e.g., votes, average_cost_for_two, etc.). Strong positive or negative correlations indicate a linear relationship between variables.
